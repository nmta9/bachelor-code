```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import CubicSpline

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from math import sqrt

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error

from statsmodels.tsa.statespace.sarimax import SARIMAX
```

## **Deep Learning Model (Bi-LSTM)**

```python
#Prepare Dataset for forecasting
class TimeSeriesDataset(Dataset):
    def __init__(self, data, seq_length=64):
        self.seq_length = seq_length
        self.series = data['Penetration'].values.astype(np.float32)

    def __len__(self):
        return len(self.series) - self.seq_length

    def __getitem__(self, idx):
        x = self.series[idx:idx+self.seq_length].reshape(-1, 1)
        y = self.series[idx + self.seq_length]
        return torch.tensor(x), torch.tensor(y)
```

```python
#Define LSTM Model
class LSTMModel(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2, dropout=0.3, bidirectional=True):
        super(LSTMModel, self).__init__()
        self.bidirectional = bidirectional
        self.lstm = nn.LSTM(
            input_size,
            hidden_size,
            num_layers,
            batch_first=True,
            dropout=dropout if num_layers > 1 else 0,
            bidirectional=bidirectional
        )
        direction_factor = 2 if bidirectional else 1
        self.fc = nn.Linear(hidden_size * direction_factor, 1)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        out, _ = self.lstm(x)
        out = out[:, -1, :]
        out = self.dropout(out)
        out = self.fc(out)
        return out.squeeze(1)
```

```python
#Preprocess data chronologically
df_weekly_lstm = df_weekly.sort_values('Date').reset_index(drop=True)
```

```python
# Split dataset by time (first 80% train, last 20% test) 
dataset_size = len(df_weekly_lstm)
train_size = int(dataset_size * 0.8)

train_df = df_weekly_lstm.iloc[:train_size].reset_index(drop=True)
test_df = df_weekly_lstm.iloc[train_size:].reset_index(drop=True)
```

```python
#Scale the target variable (only on train, then apply to test)
scaler_y = StandardScaler()
train_df['Penetration'] = scaler_y.fit_transform(train_df[['Penetration']])
test_df['Penetration'] = scaler_y.transform(test_df[['Penetration']])
```

```python
#Create train and test dataset using TimeSeriesDataset
seq_length = 64
train_dataset = TimeSeriesDataset(train_df, seq_length=seq_length)
test_dataset = TimeSeriesDataset(test_df, seq_length=seq_length)

# --------- Create DataLoaders ---------
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)
```

```python
#Print total number of sequences in train set and test set
print("Number of sequences in training set: ", len(train_loader))
print("Number of sequences in test set ", len(test_loader))
```

```python
# Initialize model, loss, optimizer
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = LSTMModel(input_size=1, hidden_size=64, num_layers=2).to(device)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
# Training loop

best_r2 = -np.inf
best_model_path = "best_lstm_model.pth"

train_losses = []
val_losses = []

epochs = 300
for epoch in range(epochs):
    model.train()
    total_train_loss = 0

    # Training
    for x_batch, y_batch in train_loader:
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        output = model(x_batch)
        loss = criterion(output, y_batch)
        loss.backward()
        optimizer.step()
        total_train_loss += loss.item() * x_batch.size(0)

    avg_train_loss = total_train_loss / len(train_loader.dataset)
    train_losses.append(avg_train_loss)

    #Validation
    model.eval()
    total_val_loss = 0
    y_true = []
    y_pred = []

    with torch.no_grad():
        for x_batch, y_batch in test_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            output = model(x_batch)
            loss = criterion(output, y_batch)
            total_val_loss += loss.item() * x_batch.size(0)

            y_true.extend(y_batch.cpu().numpy())
            y_pred.extend(output.cpu().numpy())

    avg_val_loss = total_val_loss / len(test_loader.dataset)
    val_losses.append(avg_val_loss)

    # Inverse transform to compute R2
    y_true_inv = scaler_y.inverse_transform(np.array(y_true).reshape(-1, 1)).flatten()
    y_pred_inv = scaler_y.inverse_transform(np.array(y_pred).reshape(-1, 1)).flatten()
    r2 = r2_score(y_true_inv, y_pred_inv)

    print(f"Epoch {epoch+1}/{epochs} - Train Loss: {avg_train_loss:.6f} - Val Loss: {avg_val_loss:.6f} - R2: {r2:.4f}")

    # Save best model
    if r2 > best_r2:
        best_r2 = r2
        torch.save(model.state_dict(), best_model_path)
```

```python
# Plot loss curve for lstm
plt.figure(figsize=(8, 5))
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Train vs Validation Loss")
plt.legend()
plt.grid(True)
plt.savefig("figures/lstm-loss-curve.png", bbox_inches='tight')
plt.show()
```

```python
# Evaluate LSTM model after training

best_model = LSTMModel(input_size=1, hidden_size=64, num_layers=2).to(device)
best_model.load_state_dict(torch.load(best_model_path))
best_model.to(device)
best_model.eval()

y_true_test, y_pred_test = [], []
with torch.no_grad():
    for x_batch, y_batch in test_loader:
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        output = best_model(x_batch)
        y_true_test.extend(y_batch.cpu().numpy())
        y_pred_test.extend(output.cpu().numpy())

y_true_test_inv = scaler_y.inverse_transform(np.array(y_true_test).reshape(-1, 1)).flatten()
y_pred_test_inv = scaler_y.inverse_transform(np.array(y_pred_test).reshape(-1, 1)).flatten()

lstm_rmse = np.sqrt(mean_squared_error(y_true_test_inv, y_pred_test_inv))
lstm_mae  = mean_absolute_error(y_true_test_inv, y_pred_test_inv)
lstm_mape = mean_absolute_percentage_error(y_true_test_inv, y_pred_test_inv)
lstm_r2   = r2_score(y_true_test_inv, y_pred_test_inv)

print(f"LSTM RMSE: {lstm_rmse:.4f}")
print(f"LSTM MAE : {lstm_mae:.4f}")
print(f"LSTM MAPE: {lstm_mape:.4f}")
print(f"LSTM R2  : {lstm_r2:.4f}")
```

```python
# Plot Real vs Predicted
plt.figure(figsize=(8,5))
plt.plot(y_true_test_inv, label="Real", color="blue")
plt.plot(y_pred_test_inv, label="Predicted", color="red", linestyle="--")
plt.title("Real vs Predicted - Test Set (LSTM)")
plt.xlabel("Time Steps")
plt.ylabel("Penetration")
plt.legend()
plt.grid(alpha=0.3)
plt.savefig("figures/real-vs-predicted-test-set-lstm.png", bbox_inches='tight')
plt.show()

def get_predictions(dataloader, model):
    preds, trues = [], []
    with torch.no_grad():
        for x_batch, y_batch in dataloader:
            x_batch = x_batch.to(device)
            output = model(x_batch)
            preds.extend(output.cpu().numpy())
            trues.extend(y_batch.cpu().numpy())
    return np.array(trues), np.array(preds)

y_true_train, y_pred_train = get_predictions(train_loader, best_model)
y_true_train = scaler_y.inverse_transform(y_true_train.reshape(-1, 1)).flatten()
y_pred_train = scaler_y.inverse_transform(y_pred_train.reshape(-1, 1)).flatten()

y_true_all = np.concatenate([y_true_train, y_true_test_inv])
y_pred_all = np.concatenate([y_pred_train, y_pred_test_inv])

plt.figure(figsize=(8,5))
plt.plot(range(len(y_true_all)), y_true_all, label="Real", color="blue")
plt.plot(range(len(y_pred_all)), y_pred_all, label="Predicted", color="red", linestyle="--")
plt.axvline(x=len(y_true_train), color="black", linestyle=":", label="Train/Test Split")
plt.title("Real vs Predicted - Whole Dataset (LSTM)")
plt.xlabel("Time Steps")
plt.ylabel("Penetration")
plt.legend()
plt.grid(alpha=0.3)
plt.savefig("figures/real-vs-predicted-whole-dataset-lstm.png", bbox_inches='tight')
plt.show()
```

```python
# Plot residual plot and distribution of residual plot for lstm model

y_true_test = np.array(y_true_test)
y_pred_test = np.array(y_pred_test)
residuals_test = y_true_test - y_pred_test

# Residual Plot
plt.figure(figsize=(8,5))
plt.scatter(range(len(residuals_test)), residuals_test, alpha=0.7)
plt.axhline(0, color="red", linestyle="--")
plt.title("Residual Plot (LSTM)")
plt.xlabel("Time Steps")
plt.ylabel("Residuals")
plt.savefig("figures/residuals-lstm.png", bbox_inches='tight')
plt.show()

#Distribution of Residuals
plt.figure(figsize=(8,5))
sns.histplot(residuals_test, kde=True, bins=30, color="purple")
plt.axvline(0, color="red", linestyle="--")
plt.title("Residual Distribution (LSTM)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.savefig("figures/residual-distributions-lstm.png", bbox_inches='tight')
plt.show()
```

## **Machine Learning Model 1: Random Forest**

```python
# Prepare dataset for Random Forest

df_weekly = df_weekly.sort_values('Date').reset_index(drop=True)
series = df_weekly['Penetration'].values

def create_features(series, seq_length=64):
    X, y = [], []
    for i in range(seq_length, len(series)):
        lag_window = series[i-seq_length:i].copy()
        diff_window = np.diff(lag_window, prepend=lag_window[0])
        feature_row = np.concatenate([lag_window, diff_window])
        X.append(feature_row)
        y.append(series[i])
    return np.array(X), np.array(y)

seq_length = 64
X, y = create_features(series, seq_length)

time_index = np.arange(len(series))[seq_length:].reshape(-1,1)
X = np.hstack([X, time_index])

# Train/test split (first 80% train, last 20% test)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Scale features
scaler_X = StandardScaler()
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)

scaler_y = StandardScaler()
y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1,1)).flatten()
y_test_scaled = scaler_y.transform(y_test.reshape(-1,1)).flatten()
```

```python
# Train Random Forest
rf_model = RandomForestRegressor(n_estimators=300, max_depth=15, random_state=42)
rf_model.fit(X_train_scaled, y_train_scaled)
```

```python
# Predictions
y_pred_train_scaled = rf_model.predict(X_train_scaled)
y_pred_test_scaled = rf_model.predict(X_test_scaled)

y_pred_train = scaler_y.inverse_transform(y_pred_train_scaled.reshape(-1,1)).flatten()
y_pred_test = scaler_y.inverse_transform(y_pred_test_scaled.reshape(-1,1)).flatten()

# Metrics
rf_rmse = np.sqrt(mean_squared_error(y_test, y_pred_test))
rf_mae  = mean_absolute_error(y_test, y_pred_test)
rf_mape = mean_absolute_percentage_error(y_test, y_pred_test)
rf_r2   = r2_score(y_test, y_pred_test)

print(f"Random Forest RMSE: {rf_rmse:.4f}")
print(f"Random Forest MAE : {rf_mae:.4f}")
print(f"Random Forest MAPE: {rf_mape:.4f}")
print(f"Random Forest R2  : {rf_r2:.4f}")
```

```python
#Real vs Predicted (Test Set) 
plt.figure(figsize=(8,5))
plt.plot(y_test, label="Real", color="blue")
plt.plot(y_pred_test, label="Predicted", color="red", linestyle="--")
plt.title("Real vs Predicted - Test Set (Random Forest)")
plt.xlabel("Time Steps")
plt.ylabel("Penetration")
plt.legend()
plt.savefig("figures/real-vs-predicted-test-set-rf.png", bbox_inches='tight')
plt.show()

#Real vs Predicted (Whole Dataset)
y_true_all = np.concatenate([y_train, y_test])
y_pred_all = np.concatenate([y_pred_train, y_pred_test])

plt.figure(figsize=(8,5))
plt.plot(range(len(y_true_all)), y_true_all, label="Real", color="blue")
plt.plot(range(len(y_pred_all)), y_pred_all, label="Predicted", color="red", linestyle="--")
plt.axvline(x=len(y_train), color="black", linestyle=":", label="Train/Test Split")
plt.title("Real vs Predicted - Whole Dataset (Random Forest)")
plt.xlabel("Time Steps")
plt.ylabel("Penetration")
plt.legend()
plt.savefig("figures/real-vs-predicted-whole-dataset-rf.png", bbox_inches='tight')
plt.show()
```

```python
#Residual Plot (Test Set)
residuals_test = y_test - y_pred_test
plt.figure(figsize=(8,5))
plt.scatter(range(len(residuals_test)), residuals_test, alpha=0.7)
plt.axhline(0, color="red", linestyle="--")
plt.title("Residual Plot (Random Forest)")
plt.xlabel("Time Steps")
plt.ylabel("Residuals")
plt.savefig("figures/residuals-rf.png", bbox_inches='tight')
plt.show()

#Residual Distribution (Test Set)
plt.figure(figsize=(8,5))
sns.histplot(residuals_test, kde=True, bins=30, color="purple")
plt.axvline(0, color="red", linestyle="--")
plt.title("Residual Distribution (Random Forest)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.savefig("figures/residual-distributions-rf.png", bbox_inches='tight')
plt.show()
```

## **Machine Learning Model 2: SARIMAX**

```python
#Prepare series
df_weekly = df_weekly.sort_values('Date').reset_index(drop=True)
series = df_weekly['Penetration']

#Train/Test Split
train_size = int(len(series) * 0.8)
train_series = series[:train_size]
test_series = series[train_size:]
```

```python
# Fit SARIMAX model

sarimax_model = SARIMAX(
    train_series, 
    order=(1,1,1), 
    seasonal_order=(1,1,1,52),
    enforce_stationarity=False,
    enforce_invertibility=False
)
sarimax_result = sarimax_model.fit(disp=False)

```

```python
# --------- Predictions ---------
train_pred = sarimax_result.predict(start=0, end=len(train_series)-1, dynamic=False)

# Forecast test
test_pred = sarimax_result.forecast(steps=len(test_series))

# --------- Metrics ---------
sarimax_rmse = np.sqrt(mean_squared_error(test_series, test_pred))
sarimax_mae  = mean_absolute_error(test_series, test_pred)
sarimax_mape = mean_absolute_percentage_error(test_series, test_pred)
sarimax_r2   = r2_score(test_series, test_pred)

print(f"SARIMAX RMSE: {sarimax_rmse:.4f}")
print(f"SARIMAX MAE : {sarimax_mae:.4f}")
print(f"SARIMAX MAPE: {sarimax_mape:.4f}")
print(f"SARIMAX R2  : {sarimax_r2:.4f}")
```

```python
#Plot Test Set(Actual vs Predicted) 
plt.figure(figsize=(8,5))
plt.plot(test_series.values, label="Real", color="blue")
plt.plot(test_pred.values, label="Predicted", color="red", linestyle="--")
plt.title("Real vs Predicted - Test Set (SARIMAX)")
plt.xlabel("Time Steps")
plt.ylabel("Penetration")
plt.legend()
plt.savefig("figures/real-vs-predicted-test-set-sarimax.png", bbox_inches='tight')

plt.show()

# --------- Plot Whole Series ---------
all_real = np.concatenate([train_series.values, test_series.values])
all_pred = np.concatenate([train_pred.values, test_pred.values])

plt.figure(figsize=(8,5))
plt.plot(range(len(all_real)), all_real, label="Real", color="blue")
plt.plot(range(len(all_pred)), all_pred, label="Predicted", color="red", linestyle="--")
plt.axvline(x=len(train_series), color="black", linestyle=":", label="Train/Test Split")
plt.title("Real vs Predicted - Whole Datast (SARIMAX)")
plt.xlabel("Time Steps")
plt.ylabel("Penetration")
plt.legend()
plt.savefig("figures/real-vs-predicted-whole-dataset-sarimax.png", bbox_inches='tight')
plt.show()

```

```python
#Residual Plot
residuals_test = test_series.values - test_pred.values
plt.figure(figsize=(8,5))
plt.scatter(range(len(residuals_test)), residuals_test, alpha=0.7)
plt.axhline(0, color="red", linestyle="--")
plt.title("Residual Plot SARIMAX")
plt.xlabel("Time Steps")
plt.ylabel("Residuals")
plt.savefig("figures/residuals-sarimax.png", bbox_inches='tight')
plt.show()

#Residual Distribution
plt.figure(figsize=(8,5))
sns.histplot(residuals_test, bins=30, kde=True, color="purple")
plt.axvline(0, color="red", linestyle="--")
plt.title("Residual Distribution SARIMAX")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.savefig("figures/residual-distribution-sarimax.png", bbox_inches='tight')
plt.show()
```
